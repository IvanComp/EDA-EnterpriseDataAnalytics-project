Now let's move on to the normalization of our useful data. First we modify the format of the time:

```python
df['srtDate'] = pd.to_datetime(df.srtDate, dayfirst=True, utc=True)

df
```


Now we will check if there are errors in the time stamps checking if each time stamp is unique

```python
df['srtDate'].value_counts()
```

Ok, no time stamp value is the same so there are no errors in the dates. Now we can proceed with further adjustments. 
Since we have different values ​​regarding the temperature we can round off and add the unit of measurement

```python
df = df.round({'absolute_humidity': 1, 'dew_point': 1, 'environmental_temperature': 1})
df
```
